```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('../data/raw/Software_5.json', lines=True)

df.head()
```

```python
df.info()
```

```python
df['reviewerID'].unique()
```

```python
df.isnull().sum()
```

```python
df['reviewTime'] = pd.to_datetime(df['reviewTime'])
```

```python
df.dropna(subset=['reviewText', 'summary'], inplace=True)
```

```python
df['reviewerName'].fillna('Unknown', inplace=True)
```

```python
df.drop_duplicates(subset=['reviewerID', 'asin', 'unixReviewTime'], inplace=True)
```

```python
(df.isnull().sum()/(len(df)))*100
```

```python
columns_to_drop = ['image']
df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)
```

```python
df.to_csv('../data/clean_software_data.csv', index=False)
```

```python
df.to_parquet('../data/clean_reviews.parquet', index=False)
```
